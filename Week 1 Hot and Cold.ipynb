```python
import random
import re 
```

```python
num = random.randint(1, 100)
# num = 42 
```

```python
guess = input("My guess between 1 and 100: ")
guess = re.sub(r'\D', '', guess)
guess = int(guess)



while True:    
    if guess == num:
        print("Correct! You get a cookie!")
        break 

    elif guess != num: 
        og_coldness = abs(guess - num)

        if guess < num:
            guess = input("Too low! Try again: ")
            guess = re.sub(r'\D', '', guess)
            guess = int(guess)

        elif guess > num:
            guess = input("Too high! Try again: ")
            guess = re.sub(r'\D', '', guess)
            guess = int(guess)

        new_coldness = abs(guess - num)

        print("Og Coldness" + str(og_coldness))
        print("New coldness" + str(new_coldness))


        if new_coldness > og_coldness: 
            print("You're getting colder!")

        elif new_coldness < og_coldness: 
            print("You're getting warmer!")

        elif new_coldness == og_coldness: 
            print("You're getting neither warmer nor colder!")

```
